# Listas ([]) - podemos adicionar ou remover itens

```python
# Criando uma lista de cidades para análise de dados
cidades = ["São Paulo", "Rio de Janeiro", "Belo Horizonte"]

# 1. Acessando pelo índice (lembrando que começa no 0)
print(cidades[0])  # Saída: São Paulo

# 2. Modificando um item (elas são mutáveis!)
cidades[1] = "Curitiba"

# 3. Adicionando um novo item no final
cidades.append("Salvador")

print(cidades)  
# Saída: ['São Paulo', 'Curitiba', 'Belo Horizonte', 'Salvador']
```

# Tuplas (()) - As tuplas são primas quase idênticas das listas, mas com uma diferença crucial: elas são imutáveis. Uma vez que você cria uma tuva, não pode adicionar, remover ou alterar nenhum elemento.

```python
# Criando uma tupla com as coordenadas da nossa matriz (Latitude, Longitude)
coordenadas_matriz = (-23.5505, -46.6333)

# Acessando o valor (igual à lista)
print(coordenadas_matriz[0])  # Saída: -23.5505

# SE VOCÊ TENTAR FAZER ISSO:
# coordenadas_matriz[0] = -22.9068
# O Python vai gerar um erro: TypeError: 'tuple' object does not support item assignment
```

# Dicionários ({}) - Esqueça os índices numéricos (0, 1, 2...). O dicionário funciona com o conceito de Chave e Valor (key: value). É a estrutura perfeita para representar uma linha de uma tabela de banco de dados (SQL) ou um formato JSON. Em vez de buscar o dado pela posição dele, você busca pelo "nome" (chave).

```python
# Representando os dados de um cliente de forma estruturada
cliente = {
    "id": 1024,
    "nome": "Ana Silva",
    "idade": 29,
    "cidade": "Florianópolis"
}

# 1. Acessando um valor através da chave
print(cliente["nome"])  # Saída: Ana Silva

# 2. Adicionando uma nova chave/valor (ou modificando uma existente)
cliente["status"] = "Ativo"
cliente["idade"] = 30  # Atualizou de 29 para 30

print(cliente)
# Saída: {'id': 1024, 'nome': 'Ana Silva', 'idade': 30, 'cidade': 'Florianópolis', 'status': 'Ativo'}
```

# Exemplo - Carrinho de compras

```python
# 1. Estrutura de Dados: Uma LISTA cheia de DICIONÁRIOS
# Cada dicionário representa um produto (uma "linha" do nosso carrinho)
carrinho = [
    {"produto": "Teclado Mecânico", "preco": 350.00, "quantidade": 1},
    {"produto": "Mouse Gamer", "preco": 180.00, "quantidade": 2},
    {"produto": "Monitor 24'", "preco": 899.90, "quantidade": 1},
    {"produto": "Cabo HDMI", "preco": 25.00, "quantidade": 3}
]

# Inicializamos a variável que vai acumular o valor total do carrinho
valor_total_carrinho = 0

print("--- EXTRATO DO CARRINHO ---")

# 2. O Loop 'for' para percorrer o carrinho
for item in carrinho:
    # Calculamos o subtotal deste item específico: Preço x Quantidade
    subtotal_item = item["preco"] * item["quantidade"]
    
    # Somamos o subtotal deste item ao grande total acumulado
    valor_total_carrinho += subtotal_item
    
    # Mostramos o detalhe de cada produto na tela
    print(f"Produto: {item['produto']:<18} | Qtd: {item['quantidade']} | Subtotal: R$ {subtotal_item:.2f}")

print("-" * 50)

# 3. Exibindo o resultado final, fora do loop
print(f"VALOR TOTAL A PAGAR: R$ {valor_total_carrinho:.2f}")
```

```python
# SQUAD B — Cadastro de produtos
cadastro_produtos = [
    {"nome": "Mouse", "preco": 45.0},
    {"nome": "Teclado", "preco": 89.0},
]

# 1. Criando o dicionário do novo produto
novo_produto = {"nome": "Monitor", "preco": 850.0}

# 2. Adicionando o dicionário à lista usando o .append()
cadastro_produtos.append(novo_produto)

# 3. Exibindo todos os produtos e seus respectivos preços com um loop for
print("--- PRODUTOS CADASTRADOS ---")
for produto in cadastro_produtos:
    # Acessamos os valores usando as chaves de texto do dicionário
    print(f"Produto: {produto['nome']:<8} | Preço: R$ {produto['preco']:.2f}")
```

Squad C

```python
lista_de_compras = ["arroz", "feijão", "leite", "arroz"]

lista_corrigida = []

# Passamos por cada item da lista original
for item in lista_de_compras:
    # Se o item AINDA NÃO estiver na nossa lista corrigida, nós o adicionamos
    if item not in lista_corrigida:
        lista_corrigida.append(item)

# Exibindo a lista final corrigida
print(lista_corrigida)
```

```python
# SQUAD E — Estoque
# Contexto: o relatório de valor total do estoque está saindo com um número errado.
# Missão: encontrem o bug no cálculo abaixo e corrijam para que o total considere preço x quantidade.

estoque = [
    ("Caderno", 12.5, 10),   # (produto, preco, quantidade) -- tupla
    ("Lápis", 2.0, 30),
]

total = 0
for produto, preco, quantidade in estoque:
    # CORREÇÃO: Multiplicamos o preço unitário pela quantidade em estoque
    total = total + (preco * quantidade)   

print(f"Total em estoque: R$ {total:.2f}")
```
